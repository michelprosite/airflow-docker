## Essas trigger rules são usadas para controlar a execução de tarefas em um fluxo de trabalho do Airflow, permitindo definir condições precisas para o acionamento das tarefas com base no estado de suas dependências.


* all_success: A tarefa é acionada somente se todas as suas tarefas pai foram concluídas com sucesso.

* all_failed: A tarefa é acionada somente se todas as suas tarefas pai falharam.

* all_done: A tarefa é acionada quando todas as suas tarefas pai foram concluídas, independentemente do estado (sucesso, falha ou outros estados).

* one_success: A tarefa é acionada quando pelo menos uma de suas tarefas pai é concluída com sucesso.

* one_failed: A tarefa é acionada quando pelo menos uma de suas tarefas pai falha.

* none_failed: A tarefa é acionada somente se nenhuma de suas tarefas pai falhar.

* none_skipped: A tarefa é acionada somente se nenhuma de suas tarefas pai for pulada (skipped).

* dummy: A tarefa é acionada imediatamente, independentemente do estado de suas tarefas pai.

## Ao definir uma DAG (Directed Acyclic Graph) no Apache Airflow, você pode especificar vários parâmetros. Aqui estão os principais parâmetros que podem ser definidos em uma DAG:

* dag_id: O ID exclusivo da DAG.

* description: Uma descrição opcional da DAG.

* start_date: A data e hora de início da DAG (pode ser uma data passada ou futura).

* schedule_interval: O intervalo de agendamento da DAG, no formato cron ou timedelta.

* default_args: Um dicionário com argumentos padrão a serem aplicados a todas as tarefas da DAG.

* catchup: Um valor booleano que determina se as execuções passadas da DAG serão recuperadas (catch-up) ou não.

* max_active_runs: O número máximo de execuções ativas permitidas para a DAG.

* concurrency: O número máximo de tarefas que podem ser executadas simultaneamente para a DAG.

* tags: Uma lista de tags para categorizar a DAG.

* orientation: A orientação preferida para representar o DAG no UI do Airflow.

* default_view: A visualização padrão a ser exibida no UI do Airflow quando a DAG é acessada.

* timezone: O fuso horário utilizado para agendar as execuções da DAG.

* end_date: A data e hora de término da DAG, especificando um limite para a geração de novas execuções.

* on_success_callback: Uma função ou lista de funções a serem chamadas quando todas as tarefas da DAG são executadas com sucesso.

* on_failure_callback: Uma função ou lista de funções a serem chamadas quando ocorre uma falha em qualquer uma das tarefas da DAG.

* params: Um dicionário de parâmetros personalizados que podem ser usados pela DAG ou suas tarefas.

* doc_md: Uma string que pode conter documentação adicional em formato Markdown para a DAG.

* access_control: Controle de acesso e permissões para a DAG, permitindo restringir quem pode visualizar ou executar a DAG.

* email: O endereço de e-mail do remetente usado para enviar notificações.

* email_on_failure: Um valor booleano que determina se deve ser enviado um e-mail de notificação em caso de falha de uma tarefa.

* email_on_retry: Um valor booleano que determina se deve ser enviado um e-mail de notificação ao reexecutar uma tarefa.

* email_on_success: Um valor booleano que determina se deve ser enviado um e-mail de notificação em caso de sucesso de uma tarefa.

* email_on_dag_failure: Um valor booleano que determina se deve ser enviado um e-mail de notificação em caso de falha da DAG inteira.

* email_bcc: Uma lista de endereços de e-mail que devem receber uma cópia oculta (Bcc) dos e-mails de notificação.

* email_cc: Uma lista de endereços de e-mail que devem receber uma cópia (Cc) dos e-mails de notificação.
